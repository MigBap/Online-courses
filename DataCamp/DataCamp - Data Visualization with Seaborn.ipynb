# Data Visualization with Seaborn (Python)

```python
data_folder = "C:/Users/Admin/Dropbox/Python/ficheiros/DataCamp/DataCamp - Data Visualization with Seaborn/"
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# plot 1 image
def plot_1(image1, left=0, bottom=0, right=1, top=1, wspace=0, hspace=0, format_img=".jpg"):
    image = plt.imread(data_folder + (image1+format_img))
    plt.axis('off')
    plt.imshow(image)
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    plt.show()

# plot 2 images horizontally
def plot_2(image1, image2, left=0, bottom=0, right=2.2, top=1, wspace=0, hspace=0.05, format_img=".jpg"):
    plt.subplot(1, 2, 1)
    image = plt.imread(data_folder + (image1+format_img))
    plt.axis('off')
    plt.imshow(image)
    plt.subplot(1, 2, 2)
    image = plt.imread(data_folder + (image2+format_img))
    plt.axis('off')
    plt.imshow(image)
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    plt.show()

# plot 3 images horizontally
def plot_3(image1, image2, image3, left=0, bottom=-1, right=2.7, top=1, wspace=0, hspace=0.1, format_img=".jpg"):
    plt.subplot(1, 3, 1)
    imagem = plt.imread(data_folder + (image1+format_img))
    plt.axis('off')
    plt.imshow(imagem)
    plt.subplot(1, 3, 2)
    imagem = plt.imread(data_folder + (image2+format_img))
    plt.axis('off')
    plt.imshow(imagem)
    plt.subplot(1, 3, 3)
    imagem = plt.imread(data_folder + (image3+format_img))
    plt.axis('off')
    plt.imshow(imagem)
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    plt.show()
    
# plot 4 images horizontally
def plot_4(image1, image2, image3, image4, left=0, bottom=-1, right=2.7, top=1, wspace=0, hspace=0.1, format_img=".jpg"):
    plt.subplot(1, 4, 1)
    imagem = plt.imread(data_folder + (image1+format_img))
    plt.axis('off')
    plt.imshow(imagem)
    plt.subplot(1, 4, 2)
    imagem = plt.imread(data_folder + (image2+format_img))
    plt.axis('off')
    plt.imshow(imagem)
    plt.subplot(1, 4, 3)
    imagem = plt.imread(data_folder + (image3+format_img))
    plt.axis('off')
    plt.imshow(imagem)
    plt.subplot(1, 4, 4)
    imagem = plt.imread(data_folder + (image4+format_img))
    plt.axis('off')
    plt.imshow(imagem)    
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    plt.show()
```

#### Exerc√≠cio 5 - Colors in Seaborn

```python
# Colors in Seaborn

df = pd.read_csv(data_folder + "college_datav3.csv").dropna()

# Defining a color for a plot
# Seaborn supports assigning colors to plots using matplotlib color codes
print("Define color:")
sns.set(color_codes=True)
sns.distplot(df['Tuition'], color='g')
plt.title('green')
plt.show()

# Palettes
# Seaborn uses the set_palette() function to define a palette
print("Palettes:")
j=1
for p in sns.palettes.SEABORN_PALETTES:
    plt.subplot(4, 3, j)
    sns.set_palette(p)
    sns.distplot(df['Tuition'])
    plt.title(p, fontsize=18)
    plt.xlabel('')
    j=j+1

plt.subplots_adjust(left=0, bottom=-2.7, right=3, top=1, wspace=0.3, hspace=0.3)
plt.show()
```

```python
# Displaying Palettes
print("Displaying Palettes:")
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.title(p, fontsize=16)
    plt.show()
```

```python
# Defining Custom Plattes

# - Circular colors = when the data is not ordered
sns.palplot(sns.color_palette("Paired", 12))
plt.title("Paired")
plt.show()

# - Sequential colors = when the data has a consistent range from high to low
sns.palplot(sns.color_palette("Blues", 12))
plt.title("Blues")
plt.show()

# - Diverging colors = when both the low and high values are interesting
sns.palplot(sns.color_palette("BrBG", 12))
plt.title("BrBG")
plt.show()
```

#### Exerc√≠cio 9 - Matrix plots

```python
#v√≠deo
df = pd.read_csv(data_folder + "bike_share.csv")

# Getting data in the rightformat
# - Seaborn's heatmap() function requires data to be in a grid format
# - pandas crosstab() is frequently used to manipulate the data 
df_crosstab = pd.crosstab(df["mnth"], df["weekday"], values=df["total_rentals"],aggfunc='mean').round(0)
print(df_crosstab)

# Build a heatmap
plt.subplot(1, 2, 1)
sns.heatmap(df_crosstab)
plt.title('heatmap')

# Customize a heatmap
df_crosstab = df_crosstab.astype(int)
plt.subplot(1, 2, 2)
sns.heatmap(df_crosstab, annot=True, cmap="YlGnBu", cbar=False, linewidths=.5)
plt.title('heatmap customized')
plt.subplots_adjust(left=0, bottom=0, right=2, top=1, wspace=0.3, hspace=0.3)
plt.show()

# Centering a heatmap
# - Seaborn support centering the heatmap colors on a specic value
plt.subplot(1, 2, 1)
sns.heatmap(df_crosstab, annot=True, fmt="d", cmap="YlGnBu", cbar=True, center=df_crosstab.loc[9, 6])
plt.title('centering the heatmap around a specific value (6394)')

# Plotting a correlation matrix
# - Pandas corr function calculates correlations between columns in a dataframe
# - The output can be converted to a heatmap with seaborn
plt.subplot(1, 2, 2)
sns.heatmap(df.corr())
plt.title('correlation matrix')
plt.subplots_adjust(left=0, bottom=0, right=2, top=1, wspace=0.3, hspace=0.3)
plt.show()
```

#### Exerc√≠cio 1 - Using FacetGrid, factorplot and Implot

```python
#v√≠deo

df = pd.read_csv(data_folder + "college_datav3.csv").dropna()

# FacetGrid
# - The FacetGrid is foundational for many data aware grids
# - It allows the user to control how data is distributed across columns, rows and hue
# - Once a FacetGrid is created, the plot type must be mapped to the grid

# FacetGrid Categorical Example
print("FacetGrid Categorical Example:")
g = sns.FacetGrid(df, col="HIGHDEG")
g.map(sns.boxplot, 'Tuition', order=['1', '2', '3', '4'])
plt.show()

# factorplot() --> renamed to catplot()
print("catplot(): simpler way to use a FacetGrid for categorical data")
# - The factorplot is a simpler way to use a FacetGrid for categorical data
# - Combines the facetting and mapping process into 1 function
sns.catplot(x="Tuition", data=df, col="HIGHDEG", kind='box')
plt.show()

# FacetGrid for regression
print("FacetGrid for regression:")
g = sns.FacetGrid(df, col="HIGHDEG")
g.map(plt.scatter,'Tuition','SAT_AVG_ALL')
plt.show()

# lmplot
print("lmplot: plots scatter and regression plots on a FacetGrid")
sns.lmplot(data=df, x="Tuition", y="SAT_AVG_ALL", col="HIGHDEG", fit_reg=True)
plt.show()
```

Exerc√≠cio 4

```python
# Using a lmplot

degree_ord = ['Graduate', 'Bachelors', 'Associates']

# Create a FacetGrid varying by column and columns ordered with the degree_order variable
g = sns.FacetGrid(df, col="Degree_Type", col_order=degree_ord)

# Map a scatter plot of Undergrad Population compared to PCTPELL
g.map(plt.scatter, 'UG', 'PCTPELL')
plt.subplots_adjust(left=0, bottom=-0.5, right=2, top=1, wspace=0.1, hspace=0.3)
plt.show()

############################################

# Re-create the plot above as an lmplot
sns.lmplot(data=df, x='UG', y='PCTPELL', col="Degree_Type", col_order=degree_ord)
plt.subplots_adjust(left=0, bottom=-0.5, right=2, top=1, wspace=0.1, hspace=0.3)
plt.show()
```

```python
inst_ord = ['Public', 'Private non-profit']

# Create an lmplot that has a column for Ownership, a row for Degree_Type and hue based on the WOMENONLY column
sns.lmplot(data=df, x='SAT_AVG_ALL', y='Tuition', col="Ownership", row='Degree_Type',
           row_order=['Graduate', 'Bachelors'], hue='WOMENONLY', col_order=inst_ord)

plt.subplots_adjust(left=0, bottom=-0.5, right=2, top=1, wspace=0.1, hspace=0.3)
plt.show()
```

```python
# Building a PairGrid

df = pd.read_csv(data_folder + "insurance_premiums.csv")

# Create a PairGrid with a scatter plot for fatal_collisions and premiums
g = sns.PairGrid(df, vars=["fatal_collisions", "premiums"])
g.map(plt.scatter)
plt.show()

print("==========================================================================================================")

# Create the same PairGrid but map a histogram on the diag
g = sns.PairGrid(df, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(plt.hist)
g2.map_offdiag(plt.scatter)
plt.show()
```

Exerc√≠cio 7

```python
# Using a pairplot

# Create a pairwise plot of the variables using a scatter plot
sns.pairplot(data=df, vars=["fatal_collisions", "premiums"], kind='scatter')
plt.subplots_adjust(left=0, bottom=0, right=1.2, top=1, wspace=0.1, hspace=0.3)
plt.show()

# Plot the same data but use a different color palette and color code by Region
sns.pairplot(data=df, vars=["fatal_collisions", "premiums"], kind='scatter',
             hue='Region', palette='RdBu', diag_kws={'alpha':.5})

plt.legend(bbox_to_anchor=(1, 1))

plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.1, hspace=0.3)
plt.show()
```

#### Exerc√≠cio 9 - Using JointGrid and jointplot

```python
#v√≠deo

df = pd.read_csv(data_folder + "college_datav3.csv").dropna()

from scipy import stats

print("JointGrid:")

# Basic JointGrid
g = sns.JointGrid(data=df, x="Tuition", y="ADM_RATE_ALL")
g.plot(sns.regplot, sns.distplot)
plt.show()

# Advanced JointGrid
g = sns.JointGrid(data=df, x="Tuition", y="ADM_RATE_ALL")
g = g.plot_joint(sns.kdeplot)
g = g.plot_marginals(sns.kdeplot, shade=True)
g = g.annotate(stats.pearsonr)
plt.show()
#########################################################################################################
```

```python
print("jointplot: easier to use, but provides less options on customization")

# jointplot()
sns.jointplot(data=df, x="Tuition", y="ADM_RATE_ALL", kind='hex')
plt.show()

# Customizing a jointplot
g = (sns.jointplot(x="Tuition", y="ADM_RATE_ALL", 
                   kind='scatter', 
                   xlim=(0, 25000), 
                   marginal_kws=dict(bins=15, rug=True),  # to control the structure of the histogram
                   data=df.query('UG < 2500 & Ownership == "Public"')).plot_joint(sns.kdeplot)).annotate(stats.pearsonr)

plt.show()
```

Exerc√≠cio 10

```python
# Building a JointGrid and jointplot

df = pd.read_csv(data_folder + "bike_share.csv")

sns.set_style("whitegrid")

# Build a JointGrid comparing humidity ("hum") and total_rentals
g = sns.JointGrid(x="hum", y="total_rentals", data=df,xlim=(0.1, 1.0)) 
g.plot(sns.regplot, sns.distplot)
plt.show()

# Create a jointplot similar to the JointGrid 
sns.jointplot(x="hum", y="total_rentals", kind='reg', data=df)
plt.show()
```

Exerc√≠cio 11

```python
# Jointplots and regression

# Plot temp vs. total_rentals as a regression plot
sns.jointplot(x="temp", y="total_rentals", kind='reg', data=df, order=2, xlim=(0, 1)).annotate(stats.pearsonr)
plt.show()

# Plot a jointplot showing the residuals
sns.jointplot(x="temp", y="total_rentals", kind='resid', data=df, order=2)
plt.show()

print("""Based on the residual plot and the pearson r value, there is a positive relationship between temperature and 
total_rentals.""")
```

Exerc√≠cio 12

```python
# Complex jointplots

# Create a jointplot of temp vs. casual riders
# Include a kdeplot over the scatter plot
g = (sns.jointplot(x="temp", y="casual", kind='scatter', data=df, marginal_kws=dict(bins=12, rug=True))
    .plot_joint(sns.kdeplot)).annotate(stats.pearsonr)
    
plt.show()

# Replicate the above plot but only for registered riders
print("Only for registered riders:")
g = (sns.jointplot(x="temp", y="registered",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True)).plot_joint(sns.kdeplot)).annotate(stats.pearsonr)

plt.show()
```
