```python
#vídeo
# Working with 2D arrays

# remembering numpy arrays [ slice0, slice1 ] and slicing [ start : stop : stride ]
#pixels num quadro: próximo de 0 (preto) e próximo de 1 (branco)

# Using meshgrid() :
import numpy as np
import matplotlib.pyplot as plt

u = np.linspace(-2, 2, 3)   #1D array  --> contain 3 values uniformly spaced beween -2 and +2 [quanto > for o último, + pixels]
v = np.linspace(-1, 1, 5)   #1D array  --> contain 5 values uniformly spaced beween -1 and +1

X,Y = np.meshgrid(u, v)   #2D array

print('X = \n', X)
print("")
print('Y = \n', Y)

Z = X**2/25 + Y**2/4  #fórmula para construir a meshgrid

print('Z = \n', Z)

plt.set_cmap('plasma')

plt.pcolor(Z)  #depois vemos melhor o que é isto: pcolor é pseudo-color

plt.show()  
# mostra o gráfico e quanto mais próximo de 0, azul ('plasma') e de 1, branco? ('plasma')

#--problema da orientação
```

```python
# In order to visualize two-dimensional arrays of data, it is necessary to understand how to generate and manipulate 2-D arrays.

# meshgrid() function

import numpy as np
import matplotlib.pyplot as plt

# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor() - pseudo-color
plt.pcolor(Z)
plt.savefig('sine_mesh.png') # Save the figure to 'sine_mesh.png'

plt.show()
```

#### Exercício 10 - Working with images 

```python
#vídeo
# Working with images 

# Grayscale images: rectangular 2D arrays
# Color images: typically three 2D arrays (channels)
#    RGB (Red-Green-Blue)
# Channel values:
#    0 to 1 (floating-point numbers)
#    0 to 255 (8 bit integers)

# Loading images: exemplo

img = plt.imread(data_folder + "xlim_and_ylim.png")     # read image (a cores)
print(img.shape)            #  (288, 432, 4) ---> aparece num numpy_array 3D onde a 'trailing dimension' é 4
plt.imshow(img)       # show image
plt.axis('off')       # para esconder os eixos/axis ao mostrar a imagem
plt.show()            # é ainda preciso usar o matplot comando para mostrar a imagem! (a imagem default)

# Reduction to gray-scale image  ---> há várias maneiras de pôr a preto e branco
collapsed = img.mean(axis=2)   # a forma + simples é tirar a média dos canais RGB: aplicar a np.mean a axis=2, a trailing dimension
                               # Isto pq uma vez q a cor da imagem consiste em 3 matrizes com valores entre 0 e 255, e portanto 
                               # a média é uma proxy razoável para os valores RGB qd os colapsamos para uma única escala de 
                               # intensidade para cada pixel. (?)
            
print(collapsed.shape)      # (288, 432)

plt.set_cmap('gray')
plt.imshow(collapsed, cmap='gray')
plt.axis('off')
plt.show()


# Uneven samples
uneven = collapsed[::4, ::2]  # nonuniform subsampling #[::4] refers to the 4th element --> tirar os pixéis a cada 4 linha e cada 2 coluna
print(uneven.shape)  #1/4 de 288 e 1/2 de 432
plt.imshow(uneven)   # aparece distorcida
plt.axis('off')
plt.show()

# Duas formas de resolver isto:

# 1) Adjusting aspect ratio
plt.imshow(uneven, aspect=2)   # já põe direito
plt.axis('off')
plt.show()

# 2) Adjusting extent
plt.imshow(uneven, cmap='gray', extent = (0, 432, 0, 288))    # já põe direito
plt.axis('off')
plt.show()
```

```python
#vídeo
# Visualizing Regressions - Seaborn ----> construído sob o matplotlib


# Seaborn ---> works best with Pandas DataFrames
import pandas as ps
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()

# Regression --> Grouping factors (same plot), p ex 'male' e 'female'

# Using hue=...   ---> para distinguir fatores de variáveis (p ex 'male' e 'female') a cores
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1')
plt.show()

# Using col=... Grouping factors (subplots): para pôr em gráficos diferentes, não no mesmo 
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')   # isto sem o comando subplot do matplotlib !
plt.show()


# Residual plots - Using  sns.residplot()
sns.residplot(x='total_bill', y='tip', data=tips, color='indianred')     #sns.residplot()
plt.show()

# Similar arguments as lmplot(), but more flexible
#   x,y can be arrays or strings
#   data is DataFrame (optional)
# Optional arguments (e.g., color) as in matplotlib
```

#### Exercício 7 - Visualizing univariate distributions

```python
#vídeo
# Visualizing univariate distributions

# Visualizing data

#   Univariate (one variable)
#     Visualization techniques for sampled univariate data:
#       Strip plots,  Swarm plots,  Violin plots

tips = pd.read_csv(data_folder + "tips.csv")

# 
sns.stripplot(y = tips['tip'], data=tips)                              # sns.stripplot()
plt.show()

# também se podem mostrar agrupar categorias
sns.stripplot(x=tips['day'], y=tips['tip'], data=tips)    # x is the factor column
plt.ylabel('tip ($)')
plt.show()

# para se visualizar melhor os dados, tips por exemplo, espalhar os dados pode ser mais elucidativo (qts há de 2$, de 4$, etc)
sns.stripplot(x=tips['day'], y=tips['tip'], data=tips, size=4, jitter=True)        # jitter 
plt.show()

# Swarm plot
sns.swarmplot(x=tips['day'], y=tips['tip'], data=tips)                         # sns.swarmplot() 
plt.ylabel('tip ($)')
plt.show()

# dá ainda p separar por tipo, pondo cores diferentes:              
sns.swarmplot(x=tips['day'], y=tips['tip'], data=tips, hue='sex')                   # hue=''

#e ainda a orientação, tb pode dar jeito ver de outra forma:        
sns.swarmplot(x=tips['day'], y=tips['tip'], data=tips, hue='sex', orient='h')       # orient='h'  , para horizontal


# Boxplots: mostra o mínimo, máximo, valor mediano, 1º quartil, 3º quartil, e outliers
sns.boxplot(x=tips['day'], y=tips['tip'], data=tips)                          # sns.boxplot()
plt.show()

# Violin plots: a densidade vai ser maior onde houver mais pontos
sns.violinplot(x=tips['day'], y=tips['tip'], data=tips)                       # sns.violinplot()
plt.show()

# Combining plots
sns.violinplot(x=tips['day'], y=tips['tip'], data=tips, inner=None, color='lightgray')
sns.stripplot(x=tips['day'], y=tips['tip'], data=tips, size=4, jitter=True)
plt.ylabel('tip em $')
plt.show()
```

#### Exercício 10 - Histogram equalization in images

```python
#%%script false
#vídeo --> ver este melhor depois ----> PARA MELHORAR A QUALIDADE DAS IMAGENS

# melhorar a qualidade da imagem: chama-se a esta técnica 'Histogram equalization in images'
orig = plt.imread(data_folder + 'low-contrast-moon.jpg')

plt.axis('off')
plt.imshow(orig)


# histograma
pixels = orig.flatten()                                                                     # .flatten() --->
plt.hist(pixels, bins=256, range=(0, 256), density=True, color='blue', alpha=0.3)              # normed, alpha
plt.show()

# valores mínimos e máximos do range
minval, maxval = orig.min(), orig.max()                                                     # .min(), .max()
print(minval, maxval)

################################################################
# Rescaling the image
rescaled = (255/(maxval-minval)) * (pixels - minval)     # caso não estivesse entre 0 e 255, fazíamos rescale
print(rescaled.min(), rescaled.max())                    # daria (0, 255)

plt.axis('off')
# plt.imshow(rescaled)

# Original & rescaled histograms --> o rescaled usa mais espaço da imagem embora não se note grande diferença para a original

# aqui íamos ver os dois histogramas lado a lado. 
plt.hist(orig.flatten(), bins=256, density=True, color='blue', alpha=0.2)
plt.hist(rescaled.flatten(), bins=256, density=True, color='green', alpha=0.2)
plt.legend(['original', 'rescaled'])        #notar que são strings
plt.show()

################################################################

print("====================================================================================")

# -------------> Image histogram & CDF (cumulative distribution)  ---> é necessário para tornar melhor a qualidade da imagem

# para construir a CDF:
plt.hist(pixels, bins=256, density=True, color='blue', alpha=0.3)

plt.twinx()  #para pôr a escala vertical da área da esquerda (blue) diferente da da direita (red)  ------>   plt.twinx()

orig_cdf, bins, patches = plt.hist(pixels, cumulative=True, bins=256, density=True, color='red', alpha=0.3)

plt.title('Image histogram and CDF')
plt.xlim((0,255))
plt.show()

# Equalizing intensity values
new_pixels = np.interp(pixels, bins[:-1], orig_cdf*255)                                          # np.interp()
new = new_pixels.reshape(orig.shape)                                                             # .reshape()
plt.imshow(new)
plt.axis('off')
plt.title('Equalized image')
plt.show()


# The histogram option cumulative=True permits viewing the CDF instead of the PDF.
# Notice that plt.grid('off') switches off distracting grid lines.
# The command plt.twinx() allows two plots to be overlayed sharing the x-axis but with different scales on the y-axis.
```
