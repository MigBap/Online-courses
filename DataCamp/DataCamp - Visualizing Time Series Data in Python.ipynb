```python
# vídeo

# the moving average model
co2_levels_mean = co2_levels.rolling(52).mean()

ax = co2_levels_mean.plot(figsize=(8, 3), linewidth=2, fontsize=6)
ax.set_xlabel('Date')
ax.set_ylabel('y axis values')
ax.set_title('52 weeks rolling mean of the time series')
plt.show()

# computing aggregate values of your time series
co2_levels.index 
co2_levels.index = pd.to_datetime(co2_levels.index) #convert to datetime index

print(co2_levels.index.month)
print(co2_levels.index.year)


# Plotting aggregate values of your time series
index_month = co2_levels.index.month
co2_levels_by_month = co2_levels.groupby(index_month).mean()
co2_levels_by_month.plot()
plt.title("co2_levels.groupby(co2_levels.index.month).mean()", fontsize=12)
plt.show()
```

Exercício 5

```python
# Display rolling averages

# Compute the 52 weeks rolling mean of the co2_levels DataFrame
ma = co2_levels.rolling(window=52).mean()

# Compute the 52 weeks rolling standard deviation of the co2_levels DataFrame
mstd = co2_levels.rolling(window=52).std()

# Add the upper bound (defined as rolling mean + (2 * rolling standard deviation)) column to the ma DataFrame
ma['upper'] = ma['co2'] + (2 * mstd['co2'])

# Add the lower bound column to the ma DataFrame
ma['lower'] = ma['co2'] - (2 * mstd['co2'])

# Plot the content of the ma DataFrame
ax = ma.plot(linewidth=0.8, fontsize=6, figsize=(8, 5))

# Specify labels, legend, and show the plot
ax.set_xlabel('Date', fontsize=10)
ax.set_ylabel('CO2 levels in Mauai Hawaii', fontsize=10)
ax.set_title('Rolling mean and variance of CO2 levels\nin Mauai Hawaii from 1958 to 2001', fontsize=10)
plt.show()
```

# Chapter 3 - Seasonality, Trend and Noise

#### Exercício 1 - Autocorrelation and Partial autocorrelation

```python
#vídeo

# Plotting autocorrelations
from statsmodels.graphics import tsaplots

fig = tsaplots.plot_acf(co2_levels['co2'], lags=40)

plt.show()

# autocorrelation coefficient is between -1 and 1
```

Blue shaped region: margins of uncertainty (confidence intervals):

* Values above these blue shaped regions can be interpreted as a time series having a statistically significant relationship with a lagged version of itself

```python
# Partial autocorrelation in a time series data

# contrary to autocorrelation, partial autocorrelation removes the effect of previous time points

# Plotting partial autocorrelations
from statsmodels.graphics import tsaplots

fig = tsaplots.plot_pacf(co2_levels['co2'], lags=40)

plt.subplots_adjust(left=0, bottom=0, right=1.6, top=1, wspace=0, hspace=0) #acrescentei
plt.show()
```

Exercício 2

In the field of time series analysis, autocorrelation refers to the correlation of a time series with a lagged version of itself. For example, an autocorrelation of order 3 returns the correlation between a time series and its own values lagged by 3 time points.

```python
# Autocorrelation in time series data

from statsmodels.graphics import tsaplots

# Display the autocorrelation plot of your time series
fig = tsaplots.plot_acf(co2_levels['co2'], lags=24)

# Show plot
plt.show()
```

Exercício 4

Like autocorrelation, the partial autocorrelation function (PACF) measures the correlation coefficient between a time-series and lagged versions of itself. However, it extends upon this idea by also removing the effect of previous time points. For example, a partial autocorrelation function of order 3 returns the correlation between our time series (t_1, t_2, t_3, ...) and its own values lagged by 3 time points (t_4, t_5, t_6, ...), but only after removing all effects attributable to lags 1 and 2.

```python
# Partial autocorrelation in time series data

from statsmodels.graphics import tsaplots

# Display the partial autocorrelation plot of your time series
fig = tsaplots.plot_pacf(co2_levels['co2'], lags=24)

# Show plot
plt.show()
```

#### Exercício 6 - Seasonality, trend and noise in time series data

```python
#vídeo

# time series decomposition
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Para garantir que a imagem é suficientemente grande, para dar para ver
from pylab import rcParams
rcParams['figure.figsize'] = 11, 9

# decomposition
decomposition = sm.tsa.seasonal_decompose(co2_levels['co2'])

fig = decomposition.plot()
plt.show()


# Extracting components from time series decomposition
print(dir(decomposition))  # attributes of the decomposition
print(decomposition.seasonal.head())  # seasonal component


# Seasonality component in time series
decomp_seasonal = decomposition.seasonal
ax = decomp_seasonal.plot(figsize=(14,2))
ax.set_xlabel('Date')
ax.set_ylabel('Seasonality of time series')
ax.set_title('Seasonal values of the time series')
plt.show()

# Trend component in time series
decomp_trend = decomposition.trend
ax = decomp_trend.plot(figsize=(14,2))
ax.set_xlabel('Date')
ax.set_ylabel('Trend of time series')
ax.set_title('Trend values of the time series')
plt.show()


# Noise component in time series
decomp_resid = decomposition.resid
ax = decomp_resid.plot(figsize=(14,2))
ax.set_xlabel('Date')
ax.set_ylabel('Residual of time series')
ax.set_title('Residual values of the time series')
plt.show()
```

#### Exercício 9 Find relationships between multiple time series

```python
#vídeo

# when the relationship is thought to be linear, use: Pearson's correlation coefficient
# when non-linear, use: Kendall Tau or Spearman

from scipy.stats.stats import pearsonr
from scipy.stats.stats import spearmanr
from scipy.stats.stats import kendalltau

x = [1, 2, 4, 7]
y = [1, 3, 4, 8]

print("Pearson's correlation coefficient:  ", pearsonr(x, y))
print("Spearman correlation coefficient:  ", spearmanr(x, y))
print("Kendall Tau correlation coefficient:  ", kendalltau(x, y))

################################################################################################################

# correlation matrix
corr_p = meat[['beef', 'veal', 'turkey']].corr(method='pearson')
print("\nPearson's Correlation matrix:\n", corr_p)

corr_s = meat[['beef', 'veal', 'turkey']].corr(method='spearman')
print("\nSpearman Correlation matrix:\n", corr_s)
print("=================================================================================================")

################################################################################################################

# computing correlation matrices (all) and visualize it
corr_mat = meat.corr(method='pearson')

# Heatmap
import seaborn as sns
print(" Heatmap")
sns.heatmap(corr_mat)
plt.subplots_adjust(left=0, bottom=0.5, right=0.5, top=1, wspace=0, hspace=0)
plt.show()

print("=================================================================================================")

# Clustermap
print(" Clustered heatmap --> as variáveis são dispostas em clusters para ser mais fácil visualizar")
plt.subplots_adjust(left=0, bottom=0.5, right=0.5, top=1, wspace=0, hspace=0)
sns.clustermap(corr_mat)
plt.show()
```

#### Exercício 8 - Decompose multiple time series data

```python
#vídeo

#0Python dictinaries

# Decomposing multiple time series with Python dictionaries
import statsmodels.api as sm

my_dict = {}

# Extract the names of the time series
ts_names = jobs.columns

# Run time series decomposition
for i in ts_names:
    ts_decomposition = sm.tsa.seasonal_decompose(jobs[i])
    my_dict[i] = ts_decomposition

print(my_dict)


# Further, Extract decomposition components of multiple time series
my_dict_trend = {}

for i in ts_names:
    my_dict_trend[i] = my_dict[i].trend
    
# Convert to a dataframe
trend_df = pd.DataFrame.from_dict(my_dict_trend)
trend_df.dropna()

trend_df.head()
# trend_df.plot()
```

Exercício 9

```python
# Apply time series decomposition to your dataset

# Initialize dictionary
jobs_decomp = {}

# Get the names of each time series in the DataFrame: jobs_names
jobs_names = jobs.columns

# Run time series decomposition on each time series of the DataFrame
for ts in jobs_names:
    ts_decomposition = sm.tsa.seasonal_decompose(jobs[ts])
    jobs_decomp[ts] = ts_decomposition

print(jobs_decomp)
```

Exercício 10

```python
# Visualize the seasonality of multiple time series

jobs_seasonal = {}

# Extract the seasonal values for the decomposition of each time series
for ts in jobs_names:
    jobs_seasonal[ts] = jobs_decomp[ts].seasonal  # seasonal
    
# Create a DataFrame from the jobs_seasonal dictionary: seasonality_df
seasonality_df = pd.DataFrame.from_dict(jobs_seasonal)

# Remove the label for the index
seasonality_df.index.name = None

# Create a faceted plot of the seasonality_df DataFrame
seasonality_df.plot(subplots=True, layout=(4, 4), sharey=False, fontsize=6, linewidth=0.3, figsize=(16, 16))

# Show plot
plt.show()
```

 Each time series in the jobs DataFrame have very different seasonality patterns
